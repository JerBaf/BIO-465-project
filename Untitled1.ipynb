```python
def simulate_regular_spiking_HH_neuron(input_current, simulation_time):

    """A Hodgkin-Huxley neuron implemented in Brian2.

    Args:
        input_current (TimedArray): Input current injected into the HH neuron
        simulation_time (float): Simulation time [seconds]

    Returns:
        StateMonitor: Brian2 StateMonitor with recorded fields
        ["vm", "I_e", "m", "n", "h"]
    """

    # neuron parameters
    El = -70 * b2.mV
    EK = -90 * b2.mV
    ENa = 50 * b2.mV
    gl = 0.1*b2.msiemens
    gK = 5 * b2.msiemens
    gNa = 50 * b2.msiemens
    C = 1 * b2.ufarad

    # forming HH model with differential equations
    eqs = """
    I_e = input_current(t,i) : amp
    I_leak =  gl*(vm-El) : amp
    I_Na =  gNa*m**3*h*(vm-ENa) : amp
    I_K = gK*n**4*(vm-EK) : amp
    membrane_Im = I_e - I_Na - I_K - I_leak : amp
    alphah = .128*exp(-(vm/mV+43)/18)/ms : Hz
    alpham = -.32*(vm+47*mV)/(exp(-.25*(vm/mV+47))-1)/mV/ms : Hz
    alphan = -.032*(vm+45*mV)/(exp(-.2*(vm/mV+45))-1)/mV/ms : Hz
    betah = 4./(exp(-.2*(vm/mV+20))+1)/ms : Hz
    betam = .28*(vm+20*mV)/(exp(.2*(vm/mV+20))-1)/mV/ms : Hz
    betan = .5*exp(-(vm/mV+50)/40)/ms : Hz
    dh/dt = alphah*(1-h)-betah*h : 1
    dm/dt = alpham*(1-m)-betam*m : 1
    dn/dt = alphan*(1-n)-betan*n : 1
    dvm/dt = membrane_Im/C : volt
    """

    neuron = b2.NeuronGroup(1, eqs, method="exponential_euler")

    # parameter initialization
    neuron.vm = 0
    neuron.m = 0.05
    neuron.h = 0.60
    neuron.n = 0.32
   
    # tracking parameters
    st_mon = b2.StateMonitor(neuron, ["vm", "I_e","I_leak", "I_Na", "I_K", "m", "n", "h"], record=True)

    # running the simulation
    hh_net = b2.Network(neuron)
    hh_net.add(st_mon)
    hh_net.run(simulation_time)

    return st_mon
```

```python
def simulate_adaptive_HH_neuron(input_current, simulation_time):

    """A Hodgkin-Huxley neuron implemented in Brian2.

    Args:
        input_current (TimedArray): Input current injected into the HH neuron
        simulation_time (float): Simulation time [seconds]

    Returns:
        StateMonitor: Brian2 StateMonitor with recorded fields
        ["vm", "I_e", "m", "n", "h"]
    """

    # neuron parameters
    El = -70 * b2.mV
    EK = -90 * b2.mV
    ENa = 50 * b2.mV
    gl = 0.1*b2.msiemens
    gM = 0.07 * b2.msiemens
    gK = 5 * b2.msiemens
    gNa = 50 * b2.msiemens
    C = 1 * b2.ufarad

    # forming HH model with differential equations
    eqs = """
    I_e = input_current(t,i) : amp
    I_leak =  gl*(vm-El) : amp
    I_Na =  gNa*m**3*h*(vm-ENa) : amp
    I_M = gM*p*(vm-EK) : amp
    I_K = gK*n**4*(vm-EK) : amp
    membrane_Im = I_e - I_Na - I_M - I_K - I_leak : amp
    alphah = .128*exp(-(vm/mV+43)/18)/ms : Hz
    alpham = -.32*(vm+47*mV)/(exp(-.25*(vm/mV+47))-1)/mV/ms : Hz
    alphan = -.032*(vm+45*mV)/(exp(-.2*(vm/mV+45))-1)/mV/ms : Hz
    betah = 4./(exp(-.2*(vm/mV+20))+1)/ms : Hz
    betam = .28*(vm+20*mV)/(exp(.2*(vm/mV+20))-1)/mV/ms : Hz
    betan = .5*exp(-(vm/mV+50)/40)/ms : Hz
    p_inf=1./(exp(-.1*(vm/mV+40))+1) : 1
    tau_p = 2000.*ms/(3.3*exp((vm/mV + 20)/20)+exp(-(vm/mV+20)/20)) : second
    dh/dt = alphah*(1-h)-betah*h : 1
    dm/dt = alpham*(1-m)-betam*m : 1
    dn/dt = alphan*(1-n)-betan*n : 1
    dp/dt = (p_inf-p)/tau_p : 1
    dvm/dt = membrane_Im/C : volt
    """

    neuron = b2.NeuronGroup(1, eqs, method="exponential_euler")

    # parameter initialization
    neuron.vm = 0
    neuron.m = 0.05
    neuron.h = 0.60
    neuron.n = 0.32
    neuron.p = 0

    # tracking parameters
    st_mon = b2.StateMonitor(neuron, ["vm", "I_e", "I_leak", "I_Na", "I_M", "I_K", "m", "n", "h", "p"], record=True)

    # running the simulation
    hh_net = b2.Network(neuron)
    hh_net.add(st_mon)
    hh_net.run(simulation_time)

    return st_mon
```

```python
def plot_data(state_monitor, title=None):
    """Plots the state_monitor variables ["vm", "I_e", "m", "n", "h"] vs. time.

    Args:
        state_monitor (StateMonitor): the data to plot
        title (string, optional): plot title to display
    """
    states = state_monitor.get_states()
    fig= plt.figure(figsize=(15,24))
    
    ax = fig.add_subplot(611)
    ax.plot(state_monitor.t / b2.ms, state_monitor.vm[0] / b2.mV, lw=2)
    ax.axis(xmin=0,
        xmax=np.max(state_monitor.t / b2.ms))
    ax.set_xlabel("t [ms]")
    ax.set_ylabel("v [mV]")
    ax.grid()

    ax2 = fig.add_subplot(612)

    ax2.plot(state_monitor.t / b2.ms, state_monitor.m[0] / b2.volt, "black", lw=2)
    ax2.plot(state_monitor.t / b2.ms, state_monitor.n[0] / b2.volt, "blue", lw=2)
    ax2.plot(state_monitor.t / b2.ms, state_monitor.h[0] / b2.volt, "red", lw=2)
    ax2.set_xlabel("t (ms)")
    ax2.set_ylabel("act./inact.")
    ax2.legend(("m", "n", "h"))
    ax2.axis((0,
        np.max(state_monitor.t / b2.ms),
             0,1))
    ax2.grid()
    
    ax3 = fig.add_subplot(613)
    if "p" in states.keys():
        ax3.plot(state_monitor.t / b2.ms, state_monitor.p[0] / b2.volt, "green", lw=2)
        ax3.legend(("p"))
    ax3.set_xlabel("t (ms)")
    ax3.set_ylabel("act./inact.")
    ax3.axis((0,
        np.max(state_monitor.t / b2.ms),
             0,1))
    ax3.grid()

    ax4 = fig.add_subplot(614)
    ax4.plot(state_monitor.t / b2.ms, state_monitor.I_Na[0] / b2.uamp, lw=2)
    ax4.plot(state_monitor.t / b2.ms, state_monitor.I_K[0] / b2.uamp, lw=2)
    ax4.axis((
        0,
        np.max(state_monitor.t / b2.ms),
        min(min(state_monitor.I_Na[0] / b2.uamp), min(state_monitor.I_K[0] / b2.uamp)) * 1.1,
        max(max(state_monitor.I_Na[0] / b2.uamp), max(state_monitor.I_K[0] / b2.uamp)) * 1.1
    ))
    ax4.legend(("INa","IK"))
    ax4.set_xlabel("t [ms]")
    ax4.set_ylabel("Currents [micro A]")
    ax4.grid()
     
    ax5 = fig.add_subplot(615)
    ax5.plot(state_monitor.t / b2.ms, state_monitor.I_leak[0] / b2.uamp, lw=2)
    if  "I_M" in states.keys():
        ax5.plot(state_monitor.t / b2.ms, state_monitor.I_M[0] / b2.uamp, lw=2)
        ax5.axis((
            0,
            np.max(state_monitor.t / b2.ms),
            min(min(state_monitor.I_M[0] / b2.uamp), min(state_monitor.I_leak[0] / b2.uamp)) * 1.1,
            max(max(state_monitor.I_M[0] / b2.uamp), max(state_monitor.I_leak[0] / b2.uamp)) * 1.1
        ))
        ax5.legend(("Ileak","IM"))
        ax5.set_xlabel("t [ms]")
        ax5.set_ylabel("Currents [micro A]")
        ax5.grid()
    else: 
        ax5.axis((
            0,
            np.max(state_monitor.t / b2.ms),
            min(state_monitor.I_leak[0] / b2.uamp) * 1.1,
            max(state_monitor.I_leak[0] / b2.uamp) * 1.1
        ))
        ax5.legend("I_leak")
        ax5.set_xlabel("t [ms]")
        ax5.set_ylabel("Currents [micro A]")
        ax5.grid()
    
    ax6 = fig.add_subplot(616)
    ax6.plot(state_monitor.t / b2.ms, state_monitor.I_e[0] / b2.uamp, lw=2)
    ax6.axis((
        0,
        np.max(state_monitor.t / b2.ms),
        min(state_monitor.I_e[0] / b2.uamp) * 1.1,
        max(state_monitor.I_e[0] / b2.uamp) * 1.1
    ))

    ax6.set_xlabel("t [ms]")
    ax6.set_ylabel("I [micro A]")
    ax6.grid()

    if title is not None:
        plt.suptitle(title)

    plt.show()
```
